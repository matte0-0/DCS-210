## Question 2

**Q2.** Starting with the base data pipe to a filter and remove all guests from the US (country code "USA") and find the lead_time of less than 1 day. What command does this? What information does this filter provide?

**A2.** The below command filters the base data to the desired subset. This filter focuses in on the 6174 observed bookings which are from outside the US and were booked within a day of its start.

```r
hotels %>%
  filter(country != "USA", lead_time < 1)
```

## Question 3

**Q3.** Starting with the base data pipe to a filter to display bookings with at least one child or at least one baby. What command does this?

**A3.** The below command filters the base data to the desired subset. This filter focuses in on the 9332 observed bookings which are for more than just adults.

```r
hotels %>%
  filter(children >= 1 | babies >= 1)
```

## Questions 4, 5, 6, 7

**Q4.** Do you think it's more likely to find bookings with children or babies in city hotels or resort hotels?

**A4.** I think it's more likely to find bookings with children or babies in resort hotels, because I expect families looking for vacations to go for resorts more often than cities, and I think vacations would be the most common reason for children or babies to be staying at a hotel.

**Q5.** Determine the number of bookings in resort hotels that have more than one child or one baby in the room. What command does this?

## Question 8

**Q8.** Create a frequency table of the number of `adults` in a booking. Display the results in descending order so the most common observation is on top. What is the most common number of adults in bookings in this dataset? Are there any surprising results?

**A8.** The most common number of adults in bookings in this dataset is 2, both for city and resort hotels individually and together, which is not very surprising. What is somewhat surprising is the 223 bookings in city hotels with at least one child or baby but no adults.

```r
hotels %>%
  filter(children >= 1 | babies >= 1) %>%
  count(hotel, adults, sort = TRUE)
```

## Question 9

**Q9.** Repeat the above exercise once for canceled bookings (`is_canceled` coded as 1) and once for not canceled bookings (`is_canceled` coded as 0). What does this reveal about the surprising results you spotted in the previous exercise?

**A9.** Booking with no adults were proportionally more often cancelled.

```r
hotels %>%
  filter(is_canceled == 1, children >= 1 | babies >= 1) %>%
  count(hotel, adults, sort = TRUE)
```

```r
hotels %>%
  filter(is_canceled == 0, children >= 1 | babies >= 1) %>%
  count(hotel, adults, sort = TRUE)
```

## Questions 10, 11

**Q10.** Calculate minimum, mean, median, and maximum average daily rate (`adr`) grouped by `hotel` type so that you can get these statistics separately for resort and city hotels. Which type of hotel is higher, on average?

**A10.** On average, city hotels have a higher adr than resort hotels.

**Q11.** We observe two unusual values in the summary statistics above -- a negative minimum, and a very high maximum. What types of hotels are these?

**A11.** The negative minimum adr comes from a resort hotel. The very high maximum adr comes from a city hotel.

```r
hotels %>%
  group_by(hotel) %>%
  summarize(min_adr = min(adr), max_adr = max(adr), mean_adr = mean(adr), median_adr = median(adr))
```
